```python
import json
import csv

rows = [['Threshold', 'F1 Score', 'Precision', 'Recall']]
for threshold in range(1, 255):
    matrix = {
        'water': {
            'water': 0,
            'nowater': 0
        },
        'nowater': {
            'water': 0,
            'nowater': 0
        }
    }

    with open('samples_after.json', 'r') as f:
        samples = json.load(f)

        for sample in samples:
            v = sample['value']
            if sample['water']:
                if v >= threshold:
                    matrix['water']['water'] += 1
                else:
                    matrix['water']['nowater'] += 1
            else:
                if v >= threshold:
                    matrix['nowater']['water'] += 1
                else:
                    matrix['nowater']['nowater'] += 1
    try:
        precision = matrix['water']['water'] / (matrix['water']['water'] + matrix['nowater']['water'])
        recall = matrix['water']['water'] / (matrix['water']['water'] + matrix['water']['nowater'])
        f1 = (2*(precision*recall)) / (precision + recall)
        rows.append([
            threshold,
            f1,
            precision,
            recall
        ])
    except:
        pass
    
    continue
    total = len(samples)
    total_correct = matrix['water']['water'] + matrix['nowater']['nowater']
    try:
        rows.append([
            threshold,
            (total_correct / total),
            (matrix['water']['water'] / (matrix['water']['nowater'] + matrix['water']['water'])),
            (matrix['nowater']['nowater'] / (matrix['nowater']['nowater'] + matrix['nowater']['water'])),
            (matrix['water']['water'] / (matrix['nowater']['water'] + matrix['water']['water'])),
            (matrix['nowater']['nowater'] / (matrix['water']['nowater'] + matrix['nowater']['nowater'])),
        ])
    except:
        pass

with open('threshold_accuracy.csv', 'w') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
